```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("audi.csv")
```

```python
df.head(3)
```

```python
df = df.drop(columns=["index" , "href" , "MileageRank" , "PriceRank" , "PPYRank","Score"]) #istenmeyen veriler silindi
```

```python
df.head(3)
```

```python
 df.columns = ["yil","kasa","mil","motor","ps","vites","yakit","sahibi","fiyat","ppy"]
```

```python
df.head(3)
```

```python
df.info() #motor object yani str gorunuyor bunu sayı yapmalıyım
```

```python
df["motor"] = df["motor"].str.replace("L","")#L harfini motor yerinden sildim
```

```python
df["motor"] = pd.to_numeric(df["motor"]) #tür float oldu
```

```python
df.head(3)
```

```python
df = pd.get_dummies(df,columns = ["kasa","vites","yakit"],drop_first = True)
```

```python
df.info()
```

```python
df.head(3)
```

```python
linReg = LinearRegression()
y = df[["fiyat"]]
x = df.drop("fiyat",axis=1)
model = linReg.fit(x,y)
```

```python
df.head(1)
```

```python
model.predict([[2018,35000,1.6,110,2,2500,1,1]])
```

```python
model.score(x,y)
```
